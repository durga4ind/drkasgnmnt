```python
# importing required libraries
import pandas as pd
import numpy as np
```

```python
# reading the data from file 

movies_df = pd.read_csv('my_movies.csv')
```

```python
movies_df.head(30)
```

```python
# selecting the required datafrom table
mvs = movies_df.iloc[:, 5:]
```

```python
mvs.head()
```

```python
# checking for null values

mvs.isnull().sum()

# no null values 
```

```python
# importing required modules 

#Calculating Cosine Similarity between Users  

from sklearn.metrics import pairwise_distances
from scipy.spatial.distance import cosine, correlation
```

```python
# calculating distance between users by cosine similarity

w_mvs = 1 - pairwise_distances( mvs.values, metric='cosine')
```

```python
w_mvs
```

```python
#Storing the results in a dataframe
mvs_df = pd.DataFrame(w_mvs)
```

```python
#Setting  the index and column names to user ids 
mvs_df.index = mvs.index
mvs_df.columns = mvs.index
```

```python
mvs_df.head()
```

```python
# as diagnalvalues have no importance setting them to zero
np.fill_diagonal(w_mvs, 0)
mvs_df
```

```python
#To get the most similar users we can check for the cosine values for each row
# and print its index vales

ml = mvs_df.idxmax(axis=1)
ml
```

```python
# above data showing closest user but there are more than one closest user 
# if we check the data
# so we iteratethrough the data to see all similar users
```

```python
# getting max of every in each user and string the values to ml list
ml = mvs_df.max(axis=1).tolist()
```

```python
ml
```

```python
# now we iterate the max value for each user woth others
# so that all similar users can be mapped
```

```python
# iterating and storing the result to simlar dictionary
simlar = {}
i=0
for column in mvs_df:
    print('')
    print(column, ':', end = '   ')
    for row in mvs_df:
        if ml[i]==mvs_df.iloc[column,row]:  
            print( row, end = "   ")
            simlar[column]=row
    i=i+1
```

```python
# As we can see from above user 3 has same distance with 4,5, 8 users
```
