# Import libraries 

```python
#Importing required libraries

import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
from mlxtend.preprocessing import TransactionEncoder
from matplotlib import pyplot as plt
```

```python
# loading the data to m_df 
m_df = pd.read_csv('my_movies.csv')
m_df
```

```python
# seperating the required transformed data through slicing

m_df2 = m_df.iloc[:,5:]
m_df2
```

# Apriori Algorithm 

```python
# Building the Apriori Alogrithm

frequent_movies = apriori(m_df2, min_support=0.1, use_colnames=True)
frequent_movies

# apriori will give peculiar combination with support of 0.1 
```

```python
# Creating data set containing association rules with lift atleast 1

rules = association_rules(frequent_movies, metric="lift", min_threshold=1)
rules
```

```python
# selecting top lift items from dataset 
rules3 =rules.sort_values('lift',ascending = False)[0:50]
rules3
```

```python
# Plotting  support and lift for various lift levels
fig, ax = plt.subplots()


ax.scatter((rules[rules['lift']>1]).support,(rules[rules['lift']>1]).lift, label = 'Lift 1-3')
ax.scatter((rules[rules['lift']>3]).support,(rules[rules['lift']>3]).lift, label = 'Lift >3-5')
ax.scatter((rules[rules['lift']>5]).support,(rules[rules['lift']>5]).lift, label = 'lift >5')
ax.legend()
ax.set_xlabel('Support', color = 'purple', size = 15)
ax.set_ylabel('Lift' , color = 'purple', size = 15)
ax.set_title('Support Vs Lift Scatter Plot')

    
```

```python
# Plotting lift and confidence and support

x= rules.support
y= rules.confidence
z= rules.lift


fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15,5))

ax1.scatter(x,y, color= 'r')
ax1.set_xlabel('Support')
ax1.set_ylabel('Confidence')
ax1.set_title('S vs  C', color = 'blue');
ax2.scatter(x,z, color= 'b')
ax2.set_xlabel('Support')
ax2.set_ylabel('Lift')
ax2.set_title('S vs  L', color = 'green');
ax3.scatter(y,z, color= 'g')
ax3.set_xlabel('Confidence')
ax3.set_ylabel('Lift')
ax3.set_title('C vs  L', color = 'red');
plt.show()
```
