```python
# importing required libraries
import pandas as pd
import numpy as np
from sklearn.naive_bayes import GaussianNB
from sklearn import preprocessing
```

```python
# loading the data 
sdf = pd.read_csv("salaryData_Train.csv")
```

```python
sdf.head()
```

```python
# labeling the workclass column
le =preprocessing.LabelEncoder()
sdf2 = le.fit_transform(np.array(sdf['workclass']))
```

```python
# labeling for all columns
for i in sdf:
    sdf[i]= le.fit_transform(np.array(sdf[i]))
```

```python
# labelled data
sdf.head()
```

```python
# assigning values to x and y
arr = sdf.values
X= arr[:, :-1]
Y= arr[:, -1:]
```

```python
# selecting the GaussianNB model for regression and fitting the data
reg = GaussianNB()
reg.fit(X,Y)
```

```python
# Loading the test data 
sdf_test = pd.read_csv("salaryData_Test.csv")
```

```python
sdf_test.head()
```

```python
#Labelling the test data
for i in sdf:
    sdf_test[i]= le.fit_transform(np.array(sdf_test[i]))
```

```python
# assigning the data to x and y
X_test= sdf_test.iloc[:, :-1]
y_test= sdf_test.iloc[:, -1:]
```

```python
# predicting the y using x test data
predict = reg.predict(X_test)
```

```python
# creating confusion matrix
from sklearn import metrics
print(metrics.confusion_matrix(y_test,predict))
```

```python
# printing accuracy of prediction
print(metrics.accuracy_score(y_test,predict))
```
